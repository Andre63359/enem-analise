```python
import pandas as pd
```

```python
df = pd.read_csv("enem_2020_amostrado.csv", sep=";")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.groupby("UF_RESIDENCIA")[["NU_NOTA_MT", "NU_NOTA_LC","NU_NOTA_REDACAO"]].mean().sort_values("NU_NOTA_MT", ascending=False)
```

```python
df.groupby("TP_SEXO")[["NU_NOTA_MT", "NU_NOTA_LC", "NU_NOTA_REDACAO"]].mean().sort_values("NU_NOTA_MT", ascending=False)
```

```python
df.groupby("TP_ESCOLA")[["NU_NOTA_MT", "NU_NOTA_LC", "NU_NOTA_REDACAO"]].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(df["NU_NOTA_MT"], kde=True)
plt.title("distribuição da nota de matematica")
plt.xlabel("Nota")
plt.ylabel("Frequencia")
plt.show()
```

```python
sns.histplot(df["NU_NOTA_LC"], kde=True)
plt.title("distribuição da nota de linguagens")
plt.xlabel("Nota")
plt.ylabel("Frequencia")
plt.show()
```
